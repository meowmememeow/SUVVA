# Публикация сообщения
nats pub geo_records "Test message" -s nats://localhost:4222

# Подписка на тему
nats sub geo_records -s nats://localhost:4222

# Проверка состояния JetStream
nats s report -s nats://localhost:4222

# Создание потока
nats stream add GEO_RECORDS --subjects geo_records --storage file --retention limits --max-msgs 1000000 -s nats://localhost:4222

# Публикация сообщения в JetStream
nats pub geo_records "Test message" -s nats://localhost:4222

# Подписка на поток JetStream
nats consumer add GEO_RECORDS my_consumer --ack --replay=instant -s nats://localhost:4222
nats consumer next GEO_RECORDS my_consumer -s nats://localhost:4222
